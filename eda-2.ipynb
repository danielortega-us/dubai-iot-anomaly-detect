# Forward-fill Missing Values

```python
def ffill_numeric(colname, first_val = None):
    if first_val is not None: df.loc[df.index[0], colname] = first_val
    df[colname] = pd.to_numeric(df[colname], errors = 'coerce')
    df[colname] = df[colname].ffill()

def ffill_literal(colname, first_val = None):
    if first_val is not None: df.loc[df.index[0], colname] = first_val
    df[colname] = df[colname].ffill()
```

# Linear Gradient Filling for Missing Values

```python
def lfill(colname):
    df[colname] = pd.to_numeric(df[colname], errors = 'coerce')
    df[colname] = df[colname].interpolate(method = 'linear').bfill().ffill()
```

```python
lf_cols = [
    'Extract Fan VFD Speed Feedback (percent)',
    'Space Air CO2 (parts per million)',
    'Supply Air Humidity (percent)',
    'Valve Command (percent)',
    'Supply Air Flow Velocity (liters per second)',
    'Return Air Flow Velocity (liters per second)',
    'CHW Return Temperature (celsius)',
    'Fresh Air Modulating Damper Status (percent)',
    'Off Coil Temperature (celsius)',
    'On Coil Humidity (percent)',
    'Return Air CO2 (parts per million)',
    'Supply Air Temperature (celsius)',
    'CHW Supply Temperature (celsius)',
    'Return Air Temperature After HRW (celsius)',
    'Return Air Humidity After HRW (percent)',
    'Supply Fan VFD Speed Command (percent)',
    'Return Air Humidity (percent)',
    'Supply Air Pressure (pascal)',
    'Exhaust Air Temperature (celsius)',
    'Extract Fan VFD Speed Command (percent)',
    'Fresh Air Duct Temperature (celsius)',
    'Supply Fan VFD Speed Feedback (percent)',
    'Valve Position (percent)',
    'On Coil Temperature (celsius)',
    'Return Air Temperature (celsius)',
    'Space Air VOC (parts per million)',
    'Exhaust Air Humidity (percent)',
    'Exhaust Air Damper Feedback ',
    'Supply Air Damper Status ',
    'Return Air Damper Status '
]

for colname in lf_cols:
    lfill(colname)
```

# Convert Literal Values to Numeric Values

```python
df_1 = df.copy()
column_mappings = {}

for col in df.select_dtypes(include=['object']):
    print(f"Processing {col}")
    unique_values = df[col].unique()
    mapping = {value: idx for idx, value in enumerate(unique_values)}
    column_mappings[col] = mapping
    df[col] = df[col].map(mapping)

print(column_mappings)
```

```python
def get_normalized(mode):
    norm_df = df.copy()
    
    if mode == "minmax":
        for col in norm_df.columns:
            min_value = norm_df[col].min()
            max_value = norm_df[col].max()            
            norm_df[col] = (norm_df[col] - min_value) / (max_value - min_value)
    elif mode == "zscore":
        for col in norm_df.columns:
            mean = norm_df[col].mean()
            std = norm_df[col].std()            
            norm_df[col] = (norm_df[col] - mean) / std
    else:
        print("Unsupported normalization method.")
        norm_df = None

    return norm_df
```

```python
norm_df_mm = get_normalized("minmax")
norm_df_z = get_normalized("zscore")
```
