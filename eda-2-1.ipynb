# EDA & Preprocessing

```python
df = pd.read_csv("./csv/SUS TRA B1 412 FAHU 001.csv", low_memory = False)
df = df.iloc[::-1].reset_index(drop = True)
```

### Set Date Time column as index

```python
df["Date Time"] = pd.to_datetime(df["Date Time"])
df.set_index("Date Time", inplace = True)
```

## Eliminate Invalid Records
Remove records with `Extract Fan HOA Status `: `Hand` and `Extract Fan VFD Speed Feedback (percent)` : `0`

```python
df = df[~((df['Extract Fan HOA Status '] == 'Hand') & (df['Extract Fan VFD Speed Feedback (percent)'] == 0))]
```

Remove records with `Extract Fan VFD Speed Feedback (percent)`:`0`

```python
df = df[~(df['Extract Fan VFD Speed Feedback (percent)'] == 0)]
```

## Eliminate Too Sparse Columns

```python
def is_valid(column):
    if column.dtype == 'object':  # Check for string (literal) columns
        return column.apply(lambda x: x != '' and pd.notnull(x))  # Non-empty and non-null
    else:
        return column.notnull()  # Non-null for numeric columns
```

```python
valid_percentages = df.apply(lambda col: is_valid(col).sum() / len(col) * 100)
least_valid_columns = valid_percentages[valid_percentages < 30].sort_values()
least_valid_columns
```

```python
columns_to_keep = valid_percentages[valid_percentages >= 30].index
df = df[columns_to_keep]
```

```python
df.columns
```

## Fill the First Row Values with Average Value
To ensure that all the first values are valid for all columns.

```python
for column in df.columns:
    if not is_valid(df[column]).iloc[0]:  # Check if the first row value is invalid
        if pd.api.types.is_numeric_dtype(df[column]):  # For numeric columns
            avg_value = df[column].dropna().mean()  # Calculate the average
            df.loc[df.index[0], column] = avg_value
            print("Filled First Value for", column, "As", avg_value)
        else:  # For non-numeric columns
            df.loc[df.index[0], column] = 'N/A'  # Fill with a placeholder (or handle differently)
    else:
        print("Kept First Value for", column, "As", df.iloc[0][column])
```

# Linear Gradient Filling for Missing Values

```python
def lfill(colname):
    df[colname] = pd.to_numeric(df[colname], errors = 'coerce')
    df[colname] = df[colname].interpolate(method = 'linear').bfill().ffill()
```

```python
for colname in df.columns:
    lfill(colname)
```

```python
def get_normalized(mode):
    norm_df = df.copy()
    
    if mode == "minmax":
        for col in norm_df.columns:
            min_value = norm_df[col].min()
            max_value = norm_df[col].max()            
            norm_df[col] = (norm_df[col] - min_value) / (max_value - min_value)
    elif mode == "zscore":
        for col in norm_df.columns:
            mean = norm_df[col].mean()
            std = norm_df[col].std()            
            norm_df[col] = (norm_df[col] - mean) / std
    else:
        print("Unsupported normalization method.")
        norm_df = None

    return norm_df
```

```python
norm_df_mm = get_normalized("minmax")
norm_df_z = get_normalized("zscore")
```
