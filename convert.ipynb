# Convert Excel files to CSV

```python
def excel_to_csv(excel_file, dst_dir):
    xls = pd.ExcelFile(excel_file)

    for sheet_name in xls.sheet_names:
        df = pd.read_excel(xls, sheet_name = sheet_name, header = None)
        equip = df.iloc[0][1]
        
        df.columns = df.iloc[1]        
        df = df.drop(df.index[0])
        df = df.drop(df.index[0])
        df = df.reset_index(drop = True)
        
        csv_file = os.path.join(dst_dir, f"{sheet_name}.csv")
        df.to_csv(csv_file, index = False, header = True)
        
        print(f"Saved {sheet_name} as {csv_file} for {equip}")
```

```python
def convert_dataset_to_csv():
    for subdir, dirs, files in os.walk("./dataset/"):
        for f in files:
            if f.lower().endswith('.xlsx') or f.lower().endswith('.xls'):
                excel_file = os.path.join(subdir, f)
                
                try:
                    excel_to_csv(excel_file, "./csv/")
                except Exception as e:
                    print(f"Failed to convert '{excel_file}': {e}")
```

```python
convert_dataset_to_csv()
```
